## Plotting

### Oppgave 1

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return x**3 - 2*x**2 + x + 3

x = np.linspace(-2, 3, 1000)
y = f(x)

plt.plot(x, y, label="f(x)")
plt.grid()
plt.axhline(0)
plt.axvline(0)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```

### Oppgave 2

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2 - 2

def g(x):
    return x + 1

def h(x):
    return 1/x

x = np.linspace(-3, 3, 1000)
y1 = f(x)
y2 = g(x)
y3 = h(x)

plt.plot(x, y1, label="f(x)")
plt.plot(x, y2, label="g(x)")
plt.plot(x, y3, label="h(x)")
plt.grid()
plt.axhline(0)
plt.axvline(0)
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.ylim([-20, 20])
plt.show()
```
